### Построение графиков

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Без моментов

```python
df = pd.read_csv('../data/data_0.csv', delimiter=';', index_col=False)
```

```python
for col in set(df.columns) - set(['t']):
    df[col] = df[col].str.replace('~[', '', regex=False)
    df[col] = df[col].str.replace(']', '', regex=False)
    df[col] = df[col].str.split()
```

```python
platform = pd.DataFrame(df['platform_qs'].tolist(), index=df.index, columns=['q1', 'q2', 'q3'])
platform['t'] = df['t']
```

```python
platform = platform.astype(float)
```

```python


fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20,10))

ax1.plot(platform['t'], platform['q1'])
ax1.set(xlabel='time (s)', ylabel='q1', title='')
ax1.grid()

ax2.plot(platform['t'], platform['q2'])
ax2.set(xlabel='time (s)', ylabel='q2', title='')
ax2.grid()

ax3.plot(platform['t'], platform['q3'])
ax3.set(xlabel='time (s)', ylabel='q3', title='')
ax3.grid()

plt.show()
```

### С моментами

```python
df = pd.read_csv('../data/data_1.csv', delimiter=';', index_col=False)
```

```python
df
```

```python
for col in set(df.columns) - set(['t']):
    df[col] = df[col].str.replace('~[', '', regex=False)
    df[col] = df[col].str.replace(']', '', regex=False)
    df[col] = df[col].str.split()
```

```python
platform = pd.DataFrame(df['platform_qs'].tolist(), index=df.index, columns=['q1', 'q2', 'q3'])
platform['t'] = df['t']
platform = platform.astype(float)
```

```python
removal = pd.DataFrame(df['removal_qs'].tolist(), index=df.index, columns=['q1', 'q2', 'q3'])
removal['t'] = df['t']
removal = removal.astype(float)
```

```python


fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20,10))

ax1.plot(platform['t'], platform['q1'])
ax1.set(xlabel='time (s)', ylabel='q1', title='')
ax1.grid()

ax2.plot(platform['t'], platform['q2'])
ax2.set(xlabel='time (s)', ylabel='q2', title='')
ax2.grid()

ax3.plot(platform['t'], platform['q3'])
ax3.set(xlabel='time (s)', ylabel='q3', title='')
ax3.grid()

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20,10))

ax1.plot(removal['t'], removal['q1'])
ax1.set(xlabel='time (s)', ylabel='q1', title='')
ax1.grid()

ax2.plot(removal['t'], removal['q2'])
ax2.set(xlabel='time (s)', ylabel='q2', title='')
ax2.grid()

ax3.plot(removal['t'], removal['q3'])
ax3.set(xlabel='time (s)', ylabel='q3', title='')
ax3.grid()

plt.show()
```
